```python
class Jardinero:
    def __init__(self, nombre, tamano_jardin, tipo_suelo, exposicion_solar):
        # Atributos del objeto
        self.nombre = nombre
        self.tamano = tamano_jardin
        self.tipo_suelo = tipo_suelo
        self.exposicion_solar = exposicion_solar
        self.plantas_actuales = 0  # Iniciamos con 0 plantas

    # --- Funciones (Métodos) ---

    def sembrar(self, cantidad):
        self.plantas_actuales += cantidad
        print(f"\n[🌱] {self.nombre} ha sembrado {cantidad} plantas nuevas.")
        print(f"Total en el jardín: {self.plantas_actuales} plantas.")

    def podar(self):
        if self.plantas_actuales > 0:
            print(f"\n[✂️] {self.nombre} está podando el jardín de {self.tamano}...")
            print("El jardín ahora luce impecable.")
        else:
            print(f"\n[⚠️] No hay plantas para podar en este jardín.")

    def regar(self):
        print(f"\n[💧] Regando el suelo tipo '{self.tipo_suelo}'...")
        print(f"Bajo la exposición solar '{self.exposicion_solar}', las plantas están felices.")

# --- Parte Interactiva (Creación de Objetos) ---

def crear_jardinero(numero):
    print(f"\n--- Configurando Jardinero {numero} ---")
    nombre = input(f"Nombre del jardinero {numero}: ")
    tamano = input("Tamaño del jardín (Pequeño/Mediano/Grande): ")
    suelo = input("Tipo de suelo (Arcilloso/Arenoso/Fértil): ")
    sol = input("Exposición solar (Sombra/Media/Pleno sol): ")

    return Jardinero(nombre, tamano, suelo, sol)

# Creamos los dos objetos solicitados
jardinero1 = crear_jardinero(1)
jardinero2 = crear_jardinero(2)

# Demostración de funcionalidad interactiva
print(f"\n--- ¡A trabajar, {jardinero1.nombre}! ---")
jardinero1.sembrar(5)
jardinero1.regar()

print(f"\n--- ¡A trabajar, {jardinero2.nombre}! ---")
jardinero2.podar()
jardinero2.sembrar(10)
print("Programa realizado por Anais medina NL = 36")
```
